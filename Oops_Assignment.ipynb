Q1. Write a Python program to demonstrate multiple inheritance.
1. Employee class has 3 data members EmployeeID, Gender (String), Salary and
PerformanceRating(Out of 5) of type int. It has a get() function to get these details from
the user.
2. JoiningDetail class has a data member DateOfJoining of type Date and a function
getDoJ to get the Date of joining of employees.
3. Information Class uses the marks from Employee class and the DateOfJoining date
from the JoiningDetail class to calculate the top 3 Employees based on their Ratings
and then Display, using readData, all the details on these employees in Ascending
order of their Date Of Joining.

```python
#Multiple Inheritance ('One or More parent class and child class inherit the parents classes')

class Employee :
  def get(self,EmployeeID,Gender,Salary,PerformanceRating):

    self.EmployeeID = EmployeeID
    self.Gender = Gender
    self.Salary = Salary
    self.PerformanceRating = PerformanceRating

Emp1 = Employee()
Emp1.get(100,'Male',100000,4)
Emp2 = Employee()
Emp2.get(200, 'Female', 150000,5)
Emp3 = Employee()
Emp3.get(300, 'Male', 50000,3)
Emp4 = Employee()
Emp4.get(400, 'Female', 25000,2)
Emp5 = Employee()
Emp5.get(500, 'Female', 10000,1)


class JoinDetail :
  def getDoJ (self,DateofJoining ):
    self.DateofJoining= DateofJoining

Empone= JoinDetail()
Empone.getDoJ ('01-01-2020')

Emptwo= JoinDetail()
Emptwo.getDoJ ('01-01-2016')

Empthree= JoinDetail()
Empthree.getDoJ ('01-01-2025')

Empfour= JoinDetail()
Empfour.getDoJ ('01-06-2025')

Empfive= JoinDetail()
Empfive.getDoJ ('01-06-2026')


class Information (Employee,JoinDetail) :
   def readData(self):

    employees = [Emp1, Emp2, Emp3, Emp4, Emp5]
    joindetails = [Empone, Emptwo, Empthree, Empfour, Empfive]

    # pair each employee with their join detail (same position = same person)
    paired = list(zip(employees, joindetails))

    # sort all 5 by rating, descending, take top 3
    paired.sort(key=lambda p: p[0].PerformanceRating, reverse=True)
    top3 = paired[:3]

    # now sort just those top 3 by Date of Joining, ascending
    from datetime import datetime
    top3.sort(key=lambda p: datetime.strptime(p[1].DateofJoining, "%d-%m-%Y"))

    for emp, jd in top3:
        print("Employee ID:", emp.EmployeeID)
        print("Gender:", emp.Gender)
        print("Salary:", emp.Salary)
        print("Rating:", emp.PerformanceRating)
        print("Date of Joining:", jd.DateofJoining)
        print()

info = Information()
info.readData()
```

Q3. Consider an ongoing test cricket series. Following are the names of the players and their
scores in the test1 and 2.

Test Match 1 :
Dhoni : 56 , Balaji : 94


Test Match 2 : Balaji : 80 , Dravid : 105


Calculate the highest number of runs scored by an individual cricketer in both of the matches.


Create a python function Max_Score (M) that reads a dictionary M that recognizes the player
with the highest total score. This function will return ( Top player , Total Score ) .

You can consider the Top player as String who is the highest scorer and Top score as Integer .

Input : Max_Score({‘test1’:{‘Dhoni’:56, ‘Balaji : 85}, ‘test2’:{‘Dhoni’ 87, ‘Balaji’’:200}})


Output : (‘Balaji ‘ , 200)

```python

def Max_Score(M):
    player_totals = {}

    for match in M:
        for player in M[match]:
            score = M[match][player]
            if player in player_totals:
                player_totals[player] += score
            else:
                player_totals[player] = score

    top_player = max(player_totals, key=lambda p: player_totals[p])
    top_score = player_totals[top_player]

    return (top_player, top_score)



M = {
    'test1': {'Dhoni': 56, 'Balaji': 94},
    'test2': {'Balaji': 80, 'Dravid': 105}
}

print(Max_Score(M))
```

Q4. Create a simple Card game in which there are 8 cards which are randomly chosen from a
deck. The first card is shown face up. The game asks the player to predict whether the next card
in the selection will have a higher or lower value than the currently showing card.


For example, say the card that’s shown is a 3. The player chooses “higher,” and the next card is
shown. If that card has a higher value, the player is correct. In this example, if the player had
chosen “lower,” they would have been incorrect. If the player guesses correctly, they get 20
points. If they choose incorrectly, they lose 15 points. If the next card to be turned over has the
same value as the previous card, the player is incorrect.

```python
import random


deck = list(range(1, 14)) * 4
random.shuffle(deck)
cards = deck[:8]  # pick 8

points = 0

print(f"First card: {cards[0]}")

for i in range(1, 8):
    guess = input("Higher or Lower? (h/l): ").strip().lower()
    current = cards[i - 1]
    next_card = cards[i]
    print(f"Next card: {next_card}")

    if next_card > current:
        actual = 'h'
    elif next_card < current:
        actual = 'l'
    else:
        actual = 'same'
    if actual == 'same' or guess != actual:
        points -= 15
        print(f"Incorrect! -15 points. Total: {points}")
    else:
        points += 20
        print(f"Correct! +20 points. Total: {points}")

print(f"\nGame Over! Final Score: {points}")
```

Q6. Show a basic implementation of abstraction in python using the abstract classes.

1. Create an abstract class in python.

2. Implement abstraction with the other classes and base class as abstract class.

```python
from abc import ABC, abstractmethod


class Car(ABC):

    @abstractmethod
    def engine(self):
        pass

    @abstractmethod
    def tires(self):
        pass

    @abstractmethod
    def wheels(self):
        pass


class Maruti(Car):

    def __init__(self, seats, mileage):
        self.seats = seats
        self.mileage = mileage

    def engine(self):
        print("Maruti Engine: 1.2L Petrol Engine")

    def tires(self):
        print("Maruti Tires: MRF 175/65 R14")

    def wheels(self):
        print("Maruti Wheels: 14 inch Alloy Wheels")

    def car_details(self):
        print(f"Seater: {self.seats} Seater")
        print(f"Mileage: {self.mileage} km/l")



maruti = Maruti(seats=6, mileage=22)
maruti.engine()
maruti.tires()
maruti.wheels()
maruti.car_details()
```
